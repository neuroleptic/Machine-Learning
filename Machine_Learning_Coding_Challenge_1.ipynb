# Coding challenge

Create python definitions of the following five functions:

$x^2$

$x^3$

$x^3 + x^2 - 5$

$sin(x)$

$cos(x)$



```
# LAMBDA SCHOOL
#
# MACHINE LEARNING
#
# MIT LICENSE

"""Your code here"""

import numpy as np
import matplotlib.pyplot as plt

def f1(x):
  return x**2

def f2(x):
  return x**3

def f3(x):
  return x**3 + x**2 - 5

def f4(x):
  return np.sin(x)
  
def f5(x):
  return np.cos(x)
```

Call each of your functions and demonstrate their result:

```
"""Call your functions here"""

x = np.linspace(-5,5,100)
plt.plot(x,np.vectorize(f1)(x))

plt.show()
plt.plot(x,np.vectorize(f2)(x))

plt.show()
plt.plot(x,np.vectorize(f3)(x))

plt.show()
plt.plot(x,np.vectorize(f4)(x))

plt.show()
plt.plot(x,np.vectorize(f5)(x))
```

## Bonus challenge

`numpy.vectorize` copies a function that takes a single value as argument and returns a function that takes an array of values as argument, calling the single argument version for each value in the array. Implement your own version of `numpy.vectorize`.

```
def my_vectorize(x):
  def inner_func(*args):
    return list(map(x, *args))
  return inner_func
  
y_0 = f1(1)
y_1 = my_vectorize(f2)([1,2,3,4])
print(y_1)
```
